```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
df = sklearn.datasets.load_breast_cancer()
```

```python
print(df)
```

```python
data = pd.DataFrame(df.data, columns= df.feature_names)
```

```python
data.head()
```

```python
data.columns
```

```python
data['label'] = df.target
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['label'].value_counts()
```

```python
x = data.drop(columns='label',axis=1)
y = data['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

#**GridSearchCV**

used for determining best parameters for our model

```python
#loading the model
model = SVC()
```

```python
#hyperparameter
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
#gridsearch
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
#Fitting data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
#Best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

**RandomizedSearchCV**

```python
model = SVC()
```

```python
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
random = RandomizedSearchCV(model,parameters,cv=5)
```

```python
random.fit(x,y)
```

```python
random.cv_results_
```

```python
best_params = random.best_params_
print(best_params)
```

```python
best_score = random.best_score_
print(best_score)
```

```python
result = pd.DataFrame(random.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_result
```
